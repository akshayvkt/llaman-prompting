```python
import time
```

```python
prompts_dataset_json = ''
```

```python
# Initializing input tokens count = 0
total_output_tokens = 0
total_input_tokens = 0
```

```python
# Poll the 1k examples
for batch_number in range(500//10):
    
    prompts_of_batch = ""
    
    # Take in 10 prompts
    for index,prompt_item in enumerate(df_hqprompts_truncated[batch_number*10:(batch_number+1)*10]['Prompt'].astype(str).tolist()):
        
        # Append them to prompt body
        prompts_of_batch += f"""\n{index+1}. {prompt_item}"""
        
    batch_input_tokens = anthropic.count_tokens(f"""{HUMAN_PROMPT}{prompt_head}{prompts_of_batch}{AI_PROMPT}{output_anchor}""")
    total_input_tokens += batch_input_tokens
    # Run this appended prompt to Claude
    start_time = time.time()
    response = anthropic.completions.create(
        model="claude-2.0",
        max_tokens_to_sample=50000,
        prompt=f"""{HUMAN_PROMPT}{prompt_head}{prompts_of_batch}{AI_PROMPT}{output_anchor}""",
        )
    
    #Tracking time taken
    end_time = time.time()
    elapsed_time = end_time-start_time
    
    #Assigning completion output to variable
    full_response_10 = response.completion

    #Tracking item indexes
    batch_items = [batch_number*10,(batch_number+1)*10]
    
    #Tracking number of tokens generated this batch
    batch_output_tokens = anthropic.count_tokens(full_response_10)
    
    #Tracking total number of tokens generated so far
    total_output_tokens += batch_output_tokens
    print(f"""Time taken for batch {batch_number} containing {batch_items} and {batch_output_tokens} tokens is {round(elapsed_time,1)} seconds""")
    print(full_response_10)
    
    # Append them to the string object prompts_dataset_json
    prompts_dataset_json+=full_response_10

    
with open("prompt_pairs_new.txt", "w") as file:
    file.write(prompts_dataset_json)
```
